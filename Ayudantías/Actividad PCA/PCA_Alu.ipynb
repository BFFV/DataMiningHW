### Librerías utilizadas

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

### Base de datos

Se abre la base de datos:

```python
df = pd.read_csv("Casen2017_7var.csv")
df
```

### Eliminar NaN

Se borran las tuplas que contienen Nulos, ya que generan inconsistencias en los datos:

```python
df_clean = df.dropna()
df_clean
```

### Normalizar

Se borran las tuplas cuyo nivel de estudios tenía valor 99, ya que representan 'Outliers' innecesarios. Posteriormente se normalizan los datos utilizando el método del rango para dejarlos entre 0 y 1:

```python
new_df = df_clean[df_clean.e6a != 99]
norm_df = (new_df - new_df.min()) / (new_df.max() - new_df.min())
norm_df
```

Se seleccionan las columnas de 'comuna', 'e6a' e 'ytotcorh' como las más relevantes, ya que representan datos distribuidos en intervalos continuos (no binarios) y son importantes al momento de determinar el nivel de pobreza y calidad de vida de las personas (conocimiento previo). Además, se descarta 'region', ya que la covarianza es menor en el caso de las comunas, como se puede observar en la matriz respectiva:

```python
df = norm_df.drop(['hacinamiento', 'pobreza_multi_4d', 'pobreza_multi_5d', 'Unnamed: 0'], axis=1)

print('Matriz de Covarianza \n%s' %np.cov(df.T))

df = df.drop(['region'], axis=1)
```

# Visualizar base de datos generada en scatterplot 3D

Se grafican los datos procesados anteriormente, encontrando que las personas con mayor ingreso también tuvieron más estudios que la mayoría (universitarios, postgrado, etc...):

```python
plot_3d = plt.figure(figsize=(12, 12)).gca(projection='3d')
plt.title('Before PCA')
plot_3d.scatter(df['comuna'], df['e6a'], df['ytotcorh'], c=df['e6a'])
plot_3d.set_xlabel('Comuna')
plot_3d.set_ylabel('Nivel Ed.')
plot_3d.set_zlabel('Ingreso')
plt.show()
```

# COV, vectores propios, Valores propios

Se calcula la matriz de covarianza:

```python
print('Matriz de Covarianza \n%s' %np.cov(df.T))
df
```

### Ordenar vectores y valores propios

Se obtienen los vectores y valores propios asociados, posteriormente ordenándolos de mayor a menor: 

```python
cov_matrix = np.cov(df.T)

eigen_values, eigen_vectors = np.linalg.eig(cov_matrix)

print('Vectores Propios \n%s' %eigen_vectors)
print('\nValores Propios \n%s' %eigen_values)
```

```python
# Lista con tuplas (ValorPropio, VectorPropio)
eigen_pairs = [(np.abs(eigen_values[i]), eigen_vectors[:,i]) for i in range(len(eigen_values))]

# Se ordenan de forma descendiente
eigen_pairs.sort()
eigen_pairs.reverse()

print('Valores Propios Ordenados:')
for i in eigen_pairs:
    print(i[0])
```

### Plot de vectores propios sobre base de datos 

Se muestra el gráfico junto a los vectores propios:

```python
fig = plt.figure(figsize=(12, 12))
plot_3d = fig.add_subplot(111, projection='3d')
plot_3d.scatter(df['comuna'], df['e6a'], df['ytotcorh'], c=df['e6a'])
plot_3d.quiver([0,0,0], [0,0,0], [0,0,0], eigen_pairs[0][1], eigen_pairs[1][1], eigen_pairs[2][1])
plot_3d.set_xlabel('Comuna')
plot_3d.set_ylabel('Nivel Ed.')
plot_3d.set_zlabel('Ingreso')
plt.show()
```

### Reducir dimensionalidad de base de datos y calcular matriz de covarianza

Se calcula la matriz de transformación a partir de los 2 mayores vectores propios, y luego se saca el producto matricial con la matriz de los datos para así obtener la nueva matriz de sólo 2 columnas (se eliminó 'e6a'):

```python
eigen_matrix = np.hstack((eigen_pairs[0][1].reshape(3,1),
                      eigen_pairs[1][1].reshape(3,1)))

print('Matriz de Transformación Dimensional:\n', eigen_matrix)
```

```python
pca_df = df.dot(eigen_matrix)
pca_df.columns = ['comuna','ytotcorh']
pca_df
```

Se puede observar un cierto patrón de Clustering tras aplicar PCA:

```python
plt.scatter(pca_df['comuna'], pca_df['ytotcorh'])
plt.title('After PCA')
plt.xlabel('Comuna')
plt.ylabel('Ingreso')
plt.show()
```

### Repetir proceso

Se repite el proceso para finalmente obtener que la última columna (la más relevante en el estudio) corresponde a la de las comunas:

```python
print('Matriz de Covarianza Nueva \n%s' %np.cov(pca_df.T))

cov_matrix_pca = np.cov(pca_df.T)

eigen_values_pca, eigen_vectors_pca = np.linalg.eig(cov_matrix_pca)

print('Vectores Propios \n%s' %eigen_vectors_pca)
print('\nValores Propios \n%s' %eigen_values_pca)
```

```python
# Lista con tuplas (ValorPropio, VectorPropio)
eigen_pairs_pca = [(np.abs(eigen_values_pca[i]), eigen_vectors_pca[:,i]) for i in range(len(eigen_values_pca))]

# Se ordenan de forma descendiente
eigen_pairs_pca.sort()
eigen_pairs_pca.reverse()

print('Valores Propios Ordenados:')
for i in eigen_pairs_pca:
    print(i[0])
```

```python
eigen_matrix_pca = np.hstack((eigen_pairs_pca[0][1].reshape(2,1),))

print('Matriz de Transformación Dimensional:\n', eigen_matrix_pca)
```

```python
final_df = pca_df.dot(eigen_matrix_pca)
final_df.columns = ['comuna']
final_df
```

Como conclusión, se tiene que la calidad de vida, ingresos y nivel de pobreza de las personas depende principalmente de las comunas en las que viven.
